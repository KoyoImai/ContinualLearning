```python
import matplotlib.pyplot as plt
```

```python
name = ["task1", "task2", "task3", "task4", "task5"]
num_data = [500, 500, 500, 500, 10000]

# 棒グラフの作成
plt.figure(figsize=(10, 6))
plt.bar(name, num_data, edgecolor='black')

# ラベルとタイトル
plt.xlabel('Task')
plt.ylabel('Number of Data')
# plt.title('Number of Data per Task')
plt.grid(axis='y')

plt.tight_layout()
plt.savefig("./learnable_data_cifar10.pdf")

# 表示
plt.show()
```
