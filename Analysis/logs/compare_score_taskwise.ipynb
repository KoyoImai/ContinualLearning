```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# ----------  ----------------
method_log_files = {
    'SupCon-joint': [
        "/home/kouyou/ContinualLearning/survey/Analysis/logs/supcon-joint/practice_supcon-joint_ring0_cifar10_seed1_date2025_05_01/annalyze/scorewise/offline_False/cal_score.csv",
    ],
    'CCLIS': [
        "/home/kouyou/ContinualLearning/survey/Analysis/logs/cclis/cclis_cclis_ring2000_cifar10_seed0_date2025_05_11/annalyze/scorewise/offline_False/cal_score.csv"
    ],
    'SupCon': [
        "/home/kouyou/ContinualLearning/survey/Analysis/logs/supcon-joint/practice_supcon_ring0_cifar10_seed0_date2025_05_09/annalyze/scorewise/offline_False/cal_score.csv",
    ],
}

# ---------- CSVDataFrame ----------------
all_dfs = []
for method, file_list in method_log_files.items():
    for file_path in file_list:
        df = pd.read_csv(file_path)
        df["method"] = method
        all_dfs.append(df)

combined_df = pd.concat(all_dfs, ignore_index=True)

# ----------  ----------------
output_dir = Path("task_data_plots")
output_dir.mkdir(exist_ok=True)

# ---------- task_data ----------------
for task_data in sorted(combined_df['task_data'].unique()):
    plt.figure()
    for method in combined_df['method'].unique():
        subset = combined_df[(combined_df['task_data'] == task_data) & (combined_df['method'] == method)]
        plt.plot(subset['task_model'], subset['value'], marker='o', label=method)

    plt.title(f'Task Data {task_data}')
    plt.xlabel('Task Model')
    plt.ylabel('Value')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(output_dir / f"task_data_{task_data}.png")
    plt.show()
    plt.close()
```
